```python
import pandas as pd
```

```python
#Question1: Load the CSV file (using Python)
df = pd.read_csv("/content/FireBrigadeAndAmbulanceCallOuts.csv")

#Question 2: The total number of rows and columns.
print(df.shape)
tot_columns = len(list(df))
tot_num_rows = 0
for tot_rows in open("FireBrigadeAndAmbulanceCallOuts.csv"):
  tot_num_rows +=1
print("Total number of rows: {} and total number of columns: {}".format(tot_num_rows,tot_columns)) 
df
```

```python
df.head()
```

```python
#Question 3: The number of non-null rows (by column).
print("The number of non-null rows (by column):\n{}".format(df.notnull().sum()))
```

```python
#Question 4:  the number of null values (by column).
print("The number of null rows (by column):\n{}\n".format(df.isnull().sum()))
tvalue = 0
for cvalue in df.isnull().sum():
  tvalue = tvalue+cvalue

#Question 5: The number of null values for all columns
print("The number of null values for all columns: {}\n".format(tvalue))

#Question 6: t the total number of call outs by Station Area.
print("The total number of call outs by Station Area:\n{}".format(df["Station Area"].value_counts()))
```

```python
#Question 7: The total number of call outs by Date and Station Area.
print("The total number of call outs by Date and Station Area:\n{}".format(df[["Date","Station Area"]].value_counts()))
```

```python
#Question 8: The total number of call outs by Station Area and Date where the description is either Fire Car or Fire Alarm.
total_count = df.query('Description == "Fire CAR" | Description == "Fire ALARM" ').groupby(["Station Area","Date"]).count()
print("The total number of call outs by Station Area and Date where the description is either Fire Car or Fire Alarm:\n{}".format(total_count))
```

```python
# df['AH'].dropna()
df
```

```python
# Question 9: Replace any instance of “,” (in any column) with an empty string.
df = df.replace(',',"")
#Question 10: Replace any instance of “-” (in any column) with an empty string
import numpy as np
df = df.replace('-',"")
df
```

```python
#Question 11: Drop rows for the columns (AH, MAV, CD) where at least one row value is NULL. Need to checkagain
import numpy as np
df = df.replace('',np.nan) #To make the empty values as nan so that the corresponding rows can be removed from the df['AH]

df = df.dropna(axis=0, subset=['AH','MAV','CD'])
df
```

```python
#Question 12: Drop any duplicate rows (except for the first occurrence).
df.drop_duplicates(keep = 'first' ,inplace = True)
df
```

```python
# Question 13: Output the minimum time difference between TOC and ORD.
min_time_difference = (pd.to_datetime(df['ORD']) - pd.to_datetime(df['TOC'])).min()
print("The minimum time difference between TOC and ORD: {}".format(min_time_difference))

df.rename({'Station Area':'Station_Area'}, inplace = True, axis = 1)
```
